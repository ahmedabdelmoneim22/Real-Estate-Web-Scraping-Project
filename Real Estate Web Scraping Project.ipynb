```python
header = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36",
    "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8"
}
response = requests.get("https://www.propertyfinder.eg/en/search?c=1&t=1&pf=100000&pt=2000000&fu=0&rp=m&ob=mr", headers=header)
data = response.text
soup = BeautifulSoup(data, "html.parser")
all_links= []
for a in soup.select("ul li article a"):
    if "https://www.propertyfinder.eg" in a.get("href"):
        all_links.append(a.get("href"))
all_Price = []
all_Address = []
for link in all_links:
    response = requests.get(link, headers=header)
    data = response.text
    soup = BeautifulSoup(data, "html.parser")
    Price = soup.select_one(".styles_desktop_price__text__Eb_Ti")
    all_Price.append(Price.text)
    Address = soup.select_one(".styles_desktop_subtitle__XntGT")
    all_Address.append(Address.text)
for num in range(0,len(all_links)):
    print(all_Price[num])
    print(all_Address[num])
    print(all_links[num])
import pandas as pd
my_Real_Estate_Data = {
    'Address':all_Address,
    'Price':all_Price,
    'Links':all_links
}
print(my_Real_Estate_Data)
data_Frame = pd.DataFrame(my_Real_Estate_Data)
print(data_Frame)
data_Frame.to_excel(r'C:\Users\ahmed\Desktop\realestatedata.xlsx', index=False)
import csv
# Extract the keys (column headers) from the Dictionary.
keys = my_Real_Estate_Data.keys()
# Extract the values (rows) from the Dictionary.
rows = zip(*my_Real_Estate_Data.values())
# Write the dictionary to a CSV file.
file_path = r'C:\Users\ahmed\Desktop\realestatedata.csv'
with open(file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(keys)  # Write the column Headers.
    writer.writerows(rows)  # Write the Rows.
```
